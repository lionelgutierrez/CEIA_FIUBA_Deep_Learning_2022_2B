```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

## Defino la función de calculo

```python
# Comienzo de un x0 randon
# Calculo la derivada y obtengo xi+1
# Comparo la diferencia entre xi e xi+1. Si es menor que un umbral establecido paro. Si llego al maximo de iteraciones tambien paro
# Tendre como hiperparametros:
    #1) Learning rate
    #2) Numero maximo de iteraciones
    #3) Umbral de corte

## f(x) = -(x-2)**2 + 3
## df/dx = -2x + 4

def funcion(x):
    return -(x-2)**2 + 3
    
def derivada(x):
    return -2*x + 4

def obtenerMaximo(X0=0,umbral=0.01,maxIter=1000,lr=0.01):
    iter = 0
    Xant=X0
    listaX = [X0]
    listaUmbrales = [0]
    while True:
        Xi = Xant +lr*derivada(Xant)
        dif = np.abs(Xant - Xi)
        listaX.append(Xi)
        listaUmbrales.append(dif)
        iter+=1
        if (iter>=maxIter or dif<=umbral):
            break
        Xant = Xi
        
    return Xi,listaX,listaUmbrales
```

## Calculo el valor de x máximo

```python
# Hiperparametros
umbral = 0.001
maxIter = 2000
lr = 0.05
low=-100.0
high=100.0

# Valor inicial para X
X0 = np.random.uniform(low=low, high=high)

x,listaCalculos,listaUmbrales =  obtenerMaximo(X0=X0,umbral=umbral,maxIter=maxIter,lr=lr)   
print("Valor calculado para x: {} partiendo de X0 = {} ".format(x,X0))
print("Cantidad de iteraciones: ",len(listaCalculos))
#print("Valores intermedios para x: ",listaCalculos)
#print("Umbrales intermedios para x: ",listaUmbrales)
```

## Ploteo los valores de x en cada iteracion

```python
fig, ax = plt.subplots(1, 2, figsize=(18, 5))
fig.tight_layout(pad=3)
ax[0].plot(listaUmbrales[1:], color='g', lw=2)
ax[0].set_xlabel('N° de iteraciones')
ax[0].set_ylabel('Umbral iteracion')
ax[0].set_xlim(1, len(listaUmbrales))
ax[0].set_title("Evolucion del valor de umbral entre iteraciones")
iteraciones = [i for i in range(len(listaCalculos))]
ax[1].set_title("Evolucion del valor de x por iteracion")
ax[1].set_xlabel('N° de iteraciones')
ax[1].set_ylabel('X calculado')
ax[1].scatter(iteraciones,listaCalculos)
plt.show();
```

### Ploteo la funcion contra los valores de x calculados en cad iteracion

```python
y_calculados = funcion(np.array(listaCalculos))
x_graf = np.linspace(low,high,200)
y_graf = funcion(x_graf)

fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(x_graf, y_graf)
ax.plot(listaCalculos, y_calculados, marker="o", markerfacecolor="g")
ax.set_xlabel('x')
ax.set_ylabel('f(x) = -(x-2)**2 + 3 ')
plt.show()
```

## Realizo el calculo con diferentes learning rates e iteraciones y valor de umbral fijo

```python
umbral = 0.001
lista_maxIter = [10,100,10000]
lista_lr = [0.001,0.01,0.05,0.1,0.5]

# Valor inicial para X
X0 = np.random.uniform(low=low, high=high)

plt.figure(figsize=(18,80))
cant_lineas = len(lista_maxIter)*len(lista_lr)
cant_cols = 2
i = 0
for it in lista_maxIter:
    for lr in lista_lr:
        i += 1
        plt.subplot(cant_lineas,cant_cols,i);
        x,listaCalculos,listaUmbrales =  obtenerMaximo(X0=X0,umbral=umbral,maxIter=it,lr=lr)  
        
        plt.xlabel('N° de iteraciones')
        plt.ylabel('Umbral iteracion')
        plt.xlim(1, len(listaUmbrales))
        plt.title("Evolucion del umbral con iter {} y lr  {}".format(it,lr))
        listaUmbrales[0] = 0
        plt.plot(listaUmbrales, color='g', lw=2)
        i +=1
        iteraciones = [i for i in range(len(listaCalculos))]
        plt.subplot(cant_lineas,cant_cols,i);
        plt.scatter(iteraciones,listaCalculos)
        plt.xlabel('N° de iteraciones')
        plt.ylabel('X calculado')
        plt.title("Evolucion de x por iteracion con iter {} y lr {}".format(it,lr))

        
```
