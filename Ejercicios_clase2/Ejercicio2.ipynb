```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
class Mini_batch_gradient_descent:
    
    def inicializar_param(self):
        #return np.random.uniform(low=-1.0, high=1.0)
        return np.random.rand(1)

    def __init__(self):
        self.W1 = 0
        self.W2 = 0
        self.W3 = 0
        self.lista_mse = []
        self.tip_y_hat =None
        
    def inicializar_parametros(self):
        self.W1 = self.inicializar_param()
        self.W2 = self.inicializar_param()
        self.W3 = self.inicializar_param()

    def fit(self,X,y,epochs=100,lr=0.1,batch_size=1):
        n = X.shape[0]
        self.tip_y_hat = y.dtype
        self.inicializar_parametros()
        self.lista_mse = []
        for ep in range(epochs):
            sum_mse_b = 0
            iter_b = 0
            for i in range(0, len(X), batch_size):

                end = i + batch_size if i + batch_size <= len(X) else len(X)
                batch_X = X[i: end] 
                batch_y = y[i: end] 

                #1- Calculo el forward
                y_hat_b = self.W1*(batch_X**2) + self.W2*batch_X + self.W3
                
                #2- Calculo el error
                err_b = batch_y - y_hat_b
                sum_mse_b += np.sum(err_b**2)
                iter_b += 1
                
                #3 backpropagation
                grad1 =  np.sum((-2*err_b)*(batch_X**2))/y_hat_b.size
                grad2 =  np.sum((-2*err_b)*(batch_X))/y_hat_b.size
                grad3 =  np.sum(-2*err_b)/y_hat_b.size
                
                #4 Actualizo los pesos
                self.W1 = self.W1 - lr* grad1 
                self.W2 = self.W2 - lr* grad2
                self.W3 = self.W3 - lr* grad3
                
            mse = sum_mse_b / iter_b
            self.lista_mse.append(mse)
        
    def predict(self,X):
        n = X.shape[0]
        y_hat = np.empty((n,1), dtype=self.tip_y_hat)
        for i in range(n):
            #1- Calculo el forward
            y_hat[i] = (X[i]**2)*self.W1 + X[i]*self.W2 + self.W3
        return y_hat            
```

```python
class SyntheticDataset(object):
    
    lim_inf_x = 0
    lim_sup_x = 4
    
    @staticmethod
    def __funcion(x):
        return (-(x-2)**2) + 3 + (0.2 * np.sin(16*x))
    
    def __init__(self, num_muestras=100):
        self.X = np.linspace(SyntheticDataset.lim_inf_x,SyntheticDataset.lim_sup_x,num_muestras)
        self.y = SyntheticDataset.__funcion(self.X)
        
        
```

### Generamos nuestras muestras con el dataset sintético

```python
DS = SyntheticDataset(10000)
```

### Graficamos la distribución

```python
plt.figure(figsize=(10,7))
plt.scatter(DS.X,DS.y)
plt.xlabel('X')
plt.ylabel('y')
plt.title("y = (-(x-2)**2) + 3 + (0.2 * np.sin(16*x))")
plt.show();
```

### Generamos nuestro objeto minibatch y hacemos el fit contra nuestra dataset

```python
mini_batch = Mini_batch_gradient_descent()
```

```python
epochs=100
lr=0.01
batch_size=100

mini_batch.fit(DS.X,DS.y,epochs=epochs,lr=lr,batch_size=batch_size)
```

### Grafico el MSE versus la cantidad de epochs para batch_size=100

```python
fig = plt.figure(figsize=(10,7))
ax = plt.axes()
ax.set_title("Mean Square error")
ax.set_xlabel("N° de eprochs")
ax.set_ylabel("MSE")
ax.plot(mini_batch.lista_mse, color='g', lw=2)
plt.axhline(np.min(mini_batch.lista_mse), c='b', linestyle='--')
leyenda_mse = "MSE minimo (" +str(round(np.min(mini_batch.lista_mse),3)) + ")"
plt.legend(("MSE vs N° epochs",leyenda_mse))
plt.title("MSE vs N° de eprochs")
plt.show();
```

### Graficamos la funcion real vs la función generada por nuestro modelo

```python
plt.figure(figsize=(10,7))
ax = plt.axes()
plt.scatter(DS.X,DS.y, c='b',label='Y real')
y_modelo = mini_batch.W1*(DS.X*DS.X) + mini_batch.W2*DS.X  + mini_batch.W3
plt.xlabel('X')
plt.ylabel('y')
plt.title("y real versus Y calculada con minibatch")
plt.scatter(DS.X,y_modelo,c='g' ,lw=1, label='Y calculado')
ax.legend()
plt.show();
```
